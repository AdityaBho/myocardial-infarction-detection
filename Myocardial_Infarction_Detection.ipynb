```python
# --- Setup ---
import os
import pandas as pd
import re

# --- 1. List subdirectories (patients) ---
def list_subdirectories(path):
    try:
        if not os.path.isdir(path):
            print(f"Error: The path '{path}' is not a valid directory.")
            return None

        entries = os.listdir(path)
        subdirectories = [entry for entry in entries if os.path.isdir(os.path.join(path, entry))]
        subdirectories.sort()
        return subdirectories

    except FileNotFoundError:
        print(f"Error: The path '{path}' was not found.")
        return None
    except PermissionError:
        print(f"Error: Permission denied to access '{path}'.")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

# --- Path to your dataset ---
path = "/content/ptb-diagnostic-ecg-database-1.0.0"  # change this to your folder location

subdirectories = list_subdirectories(path)
patients = pd.DataFrame(subdirectories, columns=["patient_id"])
print("Found patients:", len(patients))
display(patients.head())

# --- 2. Find .hea files in each patient folder ---
def find_hea_files_in_subdirs(base_path):
    subdirs = list_subdirectories(base_path)
    if subdirs is None:
        return None

    hea_files_by_dir = {}

    for subdir in subdirs:
        try:
            subdir_path = os.path.join(base_path, subdir)
            all_files = os.listdir(subdir_path)
            hea_files = [f for f in all_files if f.endswith('.hea') and os.path.isfile(os.path.join(subdir_path, f))]
            hea_files.sort()
            hea_files_by_dir[subdir] = hea_files
        except PermissionError:
            print(f"Warning: Permission denied for directory '{subdir}'. Skipping.")
            hea_files_by_dir[subdir] = ["--> Permission Denied"]
        except Exception as e:
            print(f"Warning: Could not process directory '{subdir}': {e}. Skipping.")
            hea_files_by_dir[subdir] = [f"--> Error: {e}"]

    return hea_files_by_dir

patients_header = find_hea_files_in_subdirs(path)

# --- 3. Extract metadata fields from .hea files ---
def extract_localization_from_files(hea_data, base_path):
    data_by_patient = {}
    if not hea_data:
        return {}

    for directory, file_list in hea_data.items():
        grouped_by_info = {}
        for file_name in file_list:
            full_path = os.path.join(base_path, directory, file_name)

            # Initialize default values
            acute_loc = former_loc = disease = age = sex = smoker = vessels = additional_diagnoses = "Not specified"

            if file_name.startswith("-->"):
                acute_loc = former_loc = disease = age = sex = smoker = vessels = additional_diagnoses = "N/A (Error)"
            else:
                try:
                    with open(full_path, 'r', encoding='utf-8') as f:
                        content = f.read()

                        # Extract metadata with regex
                        acute_loc = re.search(r'# Acute infarction \(localization\): (.*)', content)
                        acute_loc = acute_loc.group(1).strip() if acute_loc else "Not specified"

                        former_loc = re.search(r'# Former infarction \(localization\): (.*)', content)
                        former_loc = former_loc.group(1).strip() if former_loc else "Not specified"

                        disease = re.search(r'# Reason for admission: (.*)', content)
                        disease = disease.group(1).strip() if disease else "Not specified"

                        age = re.search(r'# age: (.*)', content)
                        age = age.group(1).strip() if age else "Not specified"

                        sex = re.search(r'# sex: (.*)', content)
                        sex = sex.group(1).strip() if sex else "Not specified"

                        smoker = re.search(r'# Smoker: (.*)', content)
                        smoker = smoker.group(1).strip() if smoker else "Not specified"

                        vessels = re.search(r'# Number of coronary vessels involved: (.*)', content)
                        vessels = vessels.group(1).strip() if vessels else "Not specified"

                        additional_diagnoses = re.search(r'# Additional diagnoses: (.*)', content)
                        additional_diagnoses = additional_diagnoses.group(1).strip() if additional_diagnoses else "Not specified"

                except Exception as e:
                    acute_loc = former_loc = disease = age = sex = smoker = vessels = additional_diagnoses = f"Error: {e}"

            info_tuple = (acute_loc, former_loc, disease, age, sex, smoker, vessels, additional_diagnoses)
            if info_tuple not in grouped_by_info:
                grouped_by_info[info_tuple] = []
            grouped_by_info[info_tuple].append(file_name)

        # Summarize per patient
        patient_summary_list = []
        for info, files in grouped_by_info.items():
            unpacked_info = info
            patient_summary_list.append(unpacked_info + (len(files), files))

        data_by_patient[directory] = patient_summary_list

    return data_by_patient

patients_localization = extract_localization_from_files(patients_header, path)

# --- 4. Convert to DataFrame ---
def convert_to_dataframe(patient_data):
    data_for_df = []
    for patient, summary_list in patient_data.items():
        if summary_list:
            for data_tuple in summary_list:
                (acute_loc, former_loc, disease, age, sex, smoker,
                 vessels, additional_diagnoses, count, files) = data_tuple

                data_for_df.append({
                    'patient': patient,
                    'age': age,
                    'sex': sex,
                    'disease': disease,
                    'acute_localization': acute_loc,
                    'former_localization': former_loc,
                    'additional_diagnoses': additional_diagnoses,
                    'smoker': smoker,
                    'vessels_involved': vessels,
                    'file_count': count,
                    'hea_files': ", ".join(files)
                })
        else:
            data_for_df.append({
                'patient': patient,
                'age': 'N/A',
                'sex': 'N/A',
                'disease': 'N/A',
                'acute_localization': 'N/A',
                'former_localization': 'N/A',
                'additional_diagnoses': 'N/A',
                'smoker': 'N/A',
                'vessels_involved': 'N/A',
                'file_count': 0,
                'hea_files': ''
            })
    return pd.DataFrame(data_for_df)

patient_details = convert_to_dataframe(patients_localization)
display(patient_details.head())
```

```python
import wfdb
import numpy as np
from scipy.signal import butter, filtfilt
from sklearn.preprocessing import StandardScaler
import os
```

```python
def bandpass_filter(signal, lowcut=0.5, highcut=40.0, fs=1000, order=4):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, signal)
```

```python
def extract_features(record):
    """
    Takes a wfdb record and returns numerical features (per lead).
    """
    signal = record.p_signal  # ECG data as numpy array
    features = []
    for lead in range(signal.shape[1]):
        sig = bandpass_filter(signal[:, lead], fs=record.fs)
        features.extend([
            np.mean(sig),
            np.std(sig),
            np.max(sig),
            np.min(sig),
            np.sum(sig ** 2) / len(sig)  # signal energy
        ])
    return np.array(features)
```

```python
base_path = "/content/ptb-diagnostic-ecg-database-1.0.0"
```

```python
X = []   # features
y = []   # labels (1 = MI, 0 = Healthy)

for patient_dir in os.listdir(base_path):
    patient_path = os.path.join(base_path, patient_dir)
    if not os.path.isdir(patient_path):
        continue

    for file in os.listdir(patient_path):
        if file.endswith('.hea'):
            record_name = file[:-4]  # remove '.hea'
            record_path = os.path.join(patient_path, record_name)

            try:
                record = wfdb.rdrecord(record_path)
                features = extract_features(record)
                X.append(features)

                # Assign label based on metadata or folder name
                if 'infarction' in record.comments[0].lower():
                    y.append(1)
                else:
                    y.append(0)
            except Exception as e:
                print(f"Error reading {record_name}: {e}")
```

```python
import wfdb
import os
import numpy as np
import pandas as pd

def record_to_csv(record_path, output_csv_path):
    """
    Reads a WFDB record (with .hea + .dat) and writes signal data to CSV.
    Each column is one lead; rows = time samples.
    """
    # read record (automatically reads .hea + .dat)
    rec = wfdb.rdrecord(record_path)

    # rec.p_signal is a 2D numpy array: shape = (n_samples, n_leads)
    signal = rec.p_signal
    lead_names = rec.sig_name  # e.g. ["I", "II", "V1", ...]

    # create time axis (in seconds)
    fs = rec.fs
    n = signal.shape[0]
    time = np.arange(n) / fs

    # build DataFrame
    df = pd.DataFrame(signal, columns=lead_names)
    df.insert(0, 'time', time)

    # save to CSV
    df.to_csv(output_csv_path, index=False)
    print(f"Saved {output_csv_path}")

def convert_all_in_folder(input_dir, output_dir):
    """
    Walks through all WFDB records in input_dir, converts each to CSV in output_dir.
    """
    os.makedirs(output_dir, exist_ok=True)
    # find file bases (without extension) from .hea files
    for root, dirs, files in os.walk(input_dir):
        for fname in files:
            if fname.lower().endswith('.hea'):
                rec_base = fname[:-4]  # strip .hea
                rec_path = os.path.join(root, rec_base)

                # output CSV path
                csv_fname = rec_base + '.csv'
                out_csv = os.path.join(output_dir, csv_fname)

                try:
                    record_to_csv(rec_path, out_csv)
                except Exception as e:
                    print("Error converting", rec_path, ":", e)

if __name__ == '__main__':
    input_folder = "/content/ptb-diagnostic-ecg-database-1.0.0"  # Updated path
    output_folder = "/content/ecg_csv_output" # Updated path
    convert_all_in_folder(input_folder, output_folder)
```
